# Timing Stuff

```python
%%timeit
squares = [i**2 for i in range(10)]
squares
```

```python
%%timeit
squares = list()
for i in range(10):
    squares.append(i**2)
squares
```

average time taken for cell to compute and generate output for method 1 is faster as only 956ns is taken, however, the standard deviation for it is also higher, meaning there is more spread and variation to the computation of each square than for 'for' loop.
